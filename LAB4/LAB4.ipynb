## L thuyt
Thut ton di truyn (Genetic Algorithm - GA) l mt phng php tm kim da trn c ch tin ha t nhin, ly cm hng t thuyt tin ha ca Darwin. GA c s dng  tm li gii gn ti u cho cc bi ton ti u ha phc tp.

### Cc bc chnh ca thut ton:
1. **Khi to qun th**: To ngu nhin mt tp hp cc c th (gii php tim nng).
2. **nh gi  thch nghi (Fitness Evaluation)**: Tnh gi tr hm mc tiu cho mi c th.
3. **La chn (Selection)**: Chn cc c th tt nht  lai ghp (thng s dng cc phng php nh Tournament Selection hoc Roulette Wheel).
4. **Lai ghp (Crossover)**: Kt hp hai c th  to ra c th mi.
5. **t bin (Mutation)**: Thay i ngu nhin mt s gen  duy tr tnh a dng.
6. **Lp li**: Quay li bc 2 cho n khi t iu kin dng (s th h hoc gi tr fitness mong mun).



### ng dng
- Ti u ha hm s.
- Bi ton lp lch.
- Thit k k thut, machine learning, v.v.

2. Tm gi tr ti thiu ca hm s $g(x, y) = x^2 + y^2$ trong khong $x, y \in [-5, 5]$.

Ngoi ra, chng ta s trc quan ha s thay i ca gi tr fitness qua cc th h.


## V d 1: Ti u ha hm mt bin

### Bi ton

$$
f(x) = -(x^2) + 10x + 50
$$  
t gi tr ln nht trong khong  
$$
x \in [0, 10].
$$

# M gi
```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection:
        - Chn ngu nhin tournament_size c th trong qun th
        - Chn c th c fitness cao nht lm cha m 1
    Chn cha m 2 bng Tournament Selection tng t
    
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi c gi tr trung bnh ca cha m
        Ngc li:
            Con mi l cha m 1
            
    Thc hin t bin cho con mi vi xc sut mutation_rate:
        Nu t bin:
            Gn con mi gi tr ngu nhin trong [min_val, max_val]
            
    Thm con mi vo qun th mi
    
Cp nht qun th hin ti = qun th mi
In ra c th tt nht v gi tr fitness tng ng ca th h


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Th h {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr f(x) tt nht')
    plt.title('S thay i ca f(x) qua cc th h')
    plt.show()
    
    return best_individual, best_fitness

# Chy thut ton
best_x, best_f = genetic_algorithm_example1()
print(f'\nKt qu cui cng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

## V d 2: Ti u ha hm hai bin

### Bi ton  
Tm $(x, y)$ sao cho

g(x, y) = x^2 + y^2

t gi tr nh nht trong khong

$$


### M gi 

```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection trn qun th hin ti
    Chn cha m 2 bng Tournament Selection trn qun th hin ti
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi bng trung bnh gi tr (x, y) ca cha m
        Ngc li:
            Con mi l cha m 1
    Thc hin t bin cho con mi vi xc sut mutation_rate tng bin:
        Nu t bin:
            Gn gi tr ngu nhin trong [min_val, max_val] cho bin tng ng
    Thm con mi vo qun th mi

Cp nht qun th hin ti = qun th mi
In kt qu tt nht th h

````python
# Hm mc tiu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # S dng du m v GA tm max

# Khi to qun th
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# La chn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# t bin
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thut ton di truyn
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyn v gi tr thc
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Th h {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr g(x, y) tt nht')
    plt.title('S thay i ca g(x, y) qua cc th h')
    plt.show()
    
    return best_individual, -best_fitness

# Chy thut ton
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKt qu cui cng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bi tp 1
# Thay i hm mc tiu**: Th p dng thut ton cho hm $h(x) = sin(x) + cos(x)$ trong khong $x \in [0, 2\pi]$. So snh kt qu vi nghim chnh xc.

## Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

## Mc tiu
- Ti u hm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong min:
  $$
  x \in [0, 2\pi]
  $$
- Tm gi tr **x** sao cho **h(x)** t gi tr ln nht.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  tng thut ton di truyn (GA)

1. **M ha**: Biu din gi tr \( x \) bng chui nh phn di 16 bit.  
2. **Gii m**: Chuyn t chui nh phn sang gi tr thc trong khong \([0, 2\pi]\).  
3. **Fitness**: S dng chnh gi tr \( h(x) \) lm  thch nghi.  
4. **Chn lc**: Roulette Wheel Selection.  
5. **Lai ghp**: Lai 1 im vi xc sut cao.  
6. **t bin**: Lt bit vi xc sut thp.

---


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return np.sin(x) + np.cos(x)

# M ha: Chuyn i gi tr thc thnh chui nh phn
def encode(x, bits=16, min_val=0, max_val=2 * np.pi):
    normalized_value = (x - min_val) / (max_val - min_val)
    integer_value = int(normalized_value * (2**bits - 1))
    return format(integer_value, f'0{bits}b')

# Gii m: Chuyn i chui nh phn thnh gi tr thc
def decode(binary_string, bits=16, min_val=0, max_val=2 * np.pi):
    integer_value = int(binary_string, 2)
    normalized_value = integer_value / (2**bits - 1)
    return min_val + normalized_value * (max_val - min_val)

# Tnh  thch nghi
def calculate_fitness(population):
    fitness_values = [fitness_function(decode(individual)) for individual in population]
    # m bo khng c gi tr m
    return [max(f, 0) for f in fitness_values]

# Chn lc (Roulette Wheel Selection)
def select_parents(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness == 0:  # Trnh chia cho 0
        return random.choices(population, k=2)  # Chn ngu nhin nu khng c  thch nghi

    selection_probs = [f / total_fitness for f in fitness_values]
    return np.random.choice(population, size=2, p=selection_probs)

# Lai ghp
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        crossover_point = random.randint(1, len(parent1) - 1)
        child = parent1[:crossover_point] + parent2[crossover_point:]
        return child
    return parent1

# t bin
def mutate(individual, mutation_rate=0.01):
    mutated = ''.join(bit if random.random() > mutation_rate else '1' if bit == '0' else '0' for bit in individual)
    return mutated

# Thut ton di truyn
def genetic_algorithm(pop_size=50, generations=100, bits=16, min_val=0, max_val=2 * np.pi):
    # Khi to qun th
    population = [encode(random.uniform(min_val, max_val), bits) for _ in range(pop_size)]
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = calculate_fitness(population)
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1, parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax(fitness_values)]
        best_x = decode(best_individual)
        print(f'Th h {generation + 1}: x = {best_x:.4f}, h(x) = {best_fitness:.4f}')
    
    best_fitness = max(calculate_fitness(population))
    best_individual = population[np.argmax(calculate_fitness(population))]
    best_x = decode(best_individual)
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) tt nht')
    plt.title('S thay i ca h(x) qua cc th h')
    plt.show()
    
    return best_x, best_fitness

# Chy thut ton
best_x, best_f = genetic_algorithm()
print(f'\nKt qu cui cng: x = {best_x:.4f}, h(x) = {best_f:.4f}')
````

# Bi tp 2: Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

---

## Mc tiu

Ti u hm:
$$
h(x) = \sin(x) + \cos(x)
$$
trong khong:
$$
x \in [0, 2\pi]
$$

> Tm gi tr $x$ sao cho $h(x)$ t **gi tr ln nht**.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##   tng thut ton di truyn (GA)

| Thnh phn        | Chi tit                                                                 |
|------------------|-------------------------------------------------------------------------|
| **M ha**        | Dng chui nh phn 16 bit  biu din gi tr $x \in [0, 2\pi]$         |
| **Gii m**       | nh x nh phn sang gi tr thc: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chnh l $h(x) = \sin(x) + \cos(x)$                                     |
| **Chn lc**      | Roulette Wheel Selection                                                |
| **Lai ghp**      | 1 im ct vi xc sut 0.8                                              |
| **t bin**      | Xc sut 5% (0.05)                                                       |

---

````python
# code vo y nhe
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def target_function(x):
    return np.sin(x) + np.cos(x)

# Tnh fitness - m bo lun dng bng cch cng thm 2 (v min ca sin(x)+cos(x) l -2  -1.414)
def fitness_function(x):
    return target_function(x) + 2  # Dch ln  lun >= 0

# M ha: Chuyn i gi tr thc thnh chui nh phn
def encode(x, bits=16, min_val=0, max_val=2 * np.pi):
    normalized_value = (x - min_val) / (max_val - min_val)
    integer_value = int(normalized_value * (2**bits - 1))
    return format(integer_value, f'0{bits}b')

# Gii m: Chui nh phn -> gi tr thc
def decode(binary_string, bits=16, min_val=0, max_val=2 * np.pi):
    integer_value = int(binary_string, 2)
    normalized_value = integer_value / (2**bits - 1)
    return min_val + normalized_value * (max_val - min_val)

# Tnh ton fitness cho qun th
def calculate_fitness(population):
    return [fitness_function(decode(ind)) for ind in population]

# Chn lc (Roulette Wheel Selection) -  c sa  x l trng hp c bit
def select_parents(population, fitness_values):
    total_fitness = sum(fitness_values)
    
    # X l trng hp c bit khi tng fitness = 0 hoc c gi tr m
    if total_fitness <= 0:
        return random.choices(population, k=2)
    
    selection_probs = [f/total_fitness for f in fitness_values]
    try:
        parents = np.random.choice(population, size=2, p=selection_probs)
        return parents[0], parents[1]
    except ValueError:
        return random.choices(population, k=2)

# Lai ghp 1 im vi xc sut 0.8
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1)-1)
        child = parent1[:point] + parent2[point:]
        return child
    return parent1

# t bin vi xc sut 5% (0.05)
def mutate(individual, mutation_rate=0.05):
    return ''.join(
        bit if random.random() > mutation_rate 
        else '1' if bit == '0' else '0' 
        for bit in individual
    )

# Thut ton di truyn chnh
def genetic_algorithm(pop_size=50, generations=100, bits=16, min_val=0, max_val=2*np.pi):
    # Khi to qun th ngu nhin
    population = [encode(random.uniform(min_val, max_val), bits) for _ in range(pop_size)]
    best_fitness_history = []
    
    for gen in range(generations):
        # Tnh fitness
        fitness_values = calculate_fitness(population)
        target_values = [target_function(decode(ind)) for ind in population]
        best_target = max(target_values)
        best_fitness_history.append(best_target)
        
        # To qun th mi
        new_population = []
        for _ in range(pop_size):
            # Chn lc
            parent1, parent2 = select_parents(population, fitness_values)
            # Lai ghp
            child = crossover(parent1, parent2)
            # t bin
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        
        # Hin th kt qu tt nht mi th h
        best_idx = np.argmax(target_values)
        best_x = decode(population[best_idx])
        print(f"Gen {gen+1}: x = {best_x:.4f}, h(x) = {best_target:.4f}")
    
    # Kt qu cui cng
    final_fitness = calculate_fitness(population)
    final_target = [target_function(decode(ind)) for ind in population]
    best_idx = np.argmax(final_target)
    best_x = decode(population[best_idx])
    best_h = final_target[best_idx]
    
    # V  th
    plt.plot(range(generations), best_fitness_history)
    plt.axhline(y=np.sqrt(2), color='r', linestyle='--', label='Max theoretical')
    plt.xlabel('Generation')
    plt.ylabel('Best h(x)')
    plt.title('Genetic Algorithm Performance')
    plt.legend()
    plt.show()
    
    return best_x, best_h

# Chy thut ton
best_x, best_h = genetic_algorithm()
print(f"\nTh h: x = {best_x:.4f}, h(x) = {best_h:.4f}")
print(f"Kt qu cui cng: x = {np.pi/4:.4f}, h(x) = {np.sqrt(2):.4f}")
````

# Thay i tham s thut ton di truyn v phn tch tc  hi t

**Mc tiu:** Th thay i cc tham s chnh ca thut ton di truyn gm:

- Kch thc qun th: `pop_size`
- S th h: `generations`
- T l t bin: `mutation_rate`
- T l lai ghp: `crossover_rate`

v phn tch nh hng ca tng tham s n tc  hi t ca thut ton.

---

## Cc tham s v vai tr

| Tham s         |  ngha                                  | nh hng chnh                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | S c th trong qun th                 | Qun th ln gip khm ph khng gian tm kim tt hn nhng tn nhiu thi gian tnh ton. Qun th nh hi t nhanh nhng d b kt cc tr cc b. |
| `generations`   | S vng lp ca thut ton               | S th h cng nhiu th thut ton c nhiu c hi ci thin nghim, tuy nhin thi gian chy cng tng. |
| `mutation_rate` | Xc sut t bin mi bit                | t bin cao gip tng a dng gene, trnh b kt trong cc tr cc b, nhng nu qu cao lm qu trnh tm kim nhiu lon. |
| `crossover_rate`| Xc sut lai ghp gia hai c th       | Lai ghp gip phi hp gen tt, tng kh nng tm nghim tt. Nu qu thp, qun th km a dng; qu cao c th ph hy cu trc tt ca c th. |

---

## Phn tch nh hng

- **`pop_size`**: Kch thc qun th ln gip tm kim ton cc tt hn nhng chm hn, kch thc nh lm thut ton hi t nhanh nhng d mc kt.
- **`generations`**: S th h ln tng kh nng ci tin kt qu nhng lm tng thi gian tnh ton.
- **`mutation_rate`**: T l t bin va phi gip duy tr a dng di truyn, trnh kt cc b; qu cao hoc qu thp u lm gim hiu qu.
- **`crossover_rate`**: T l lai ghp hp l gip khai thc tt thng tin di truyn, tng hiu qu hi t.

````python
# code vo y 
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def target_function(x):
    return np.sin(x) + np.cos(x)

# Tnh fitness - m bo lun dng bng cch cng thm 2
def fitness_function(x):
    return target_function(x) + 2

# M ha: Chuyn i gi tr thc thnh chui nh phn
def encode(x, bits=16, min_val=0, max_val=2 * np.pi):
    normalized_value = (x - min_val) / (max_val - min_val)
    integer_value = int(normalized_value * (2**bits - 1))
    return format(integer_value, f'0{bits}b')

# Gii m: Chui nh phn -> gi tr thc
def decode(binary_string, bits=16, min_val=0, max_val=2 * np.pi):
    integer_value = int(binary_string, 2)
    normalized_value = integer_value / (2**bits - 1)
    return min_val + normalized_value * (max_val - min_val)

# Tnh ton fitness cho qun th
def calculate_fitness(population):
    return [fitness_function(decode(ind)) for ind in population]

# Chn lc (Roulette Wheel Selection)
def select_parents(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness <= 0:
        return random.choices(population, k=2)
    
    selection_probs = [f / total_fitness for f in fitness_values]
    return np.random.choice(population, size=2, p=selection_probs)

# Lai ghp 1 im vi xc sut crossover_rate
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        child = parent1[:point] + parent2[point:]
        return child
    return parent1

# t bin vi xc sut mutation_rate
def mutate(individual, mutation_rate=0.05):
    return ''.join(
        bit if random.random() > mutation_rate 
        else '1' if bit == '0' else '0' 
        for bit in individual
    )

# Thut ton di truyn chnh
def genetic_algorithm(pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8, bits=16, min_val=0, max_val=2*np.pi):
    population = [encode(random.uniform(min_val, max_val), bits) for _ in range(pop_size)]
    best_fitness_history = []
    
    for gen in range(generations):
        fitness_values = calculate_fitness(population)
        best_target = max(fitness_values)
        best_fitness_history.append(best_target)
        
        new_population = []
        for _ in range(pop_size):
            parent1, parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate)
            new_population.append(child)
        
        population = new_population
        
        # Hin th kt qu tt nht mi th h
        best_idx = np.argmax(fitness_values)
        best_x = decode(population[best_idx])
        print(f"Th h {gen + 1}: x = {best_x:.4f}, h(x) = {best_target:.4f}")
    
    return best_fitness_history

# Phn tch nh hng ca cc tham s
def analyze_parameters():
    # Thay i kch thc qun th
    pop_sizes = [10, 50, 100]
    generations = 100
    mutation_rate = 0.05
    crossover_rate = 0.8
    
    for pop_size in pop_sizes:
        print(f"\nKch thc qun th: {pop_size}")
        fitness_history = genetic_algorithm(pop_size=pop_size, generations=generations, mutation_rate=mutation_rate, crossover_rate=crossover_rate)
        plt.plot(fitness_history, label=f'pop_size={pop_size}')
    
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) tt nht')
    plt.title('nh hng ca kch thc qun th n tc  hi t')
    plt.legend()
    plt.show()

    # Thay i t l t bin
    mutation_rates = [0.01, 0.05, 0.1]
    
    for mutation_rate in mutation_rates:
        print(f"\nT l t bin: {mutation_rate}")
        fitness_history = genetic_algorithm(pop_size=50, generations=generations, mutation_rate=mutation_rate, crossover_rate=crossover_rate)
        plt.plot(fitness_history, label=f'mutation_rate={mutation_rate}')
    
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) tt nht')
    plt.title('nh hng ca t l t bin n tc  hi t')
    plt.legend()
    plt.show()

    # Thay i t l lai ghp
    crossover_rates = [0.6, 0.8, 1.0]
    
    for crossover_rate in crossover_rates:
        print(f"\nT l lai ghp: {crossover_rate}")
        fitness_history = genetic_algorithm(pop_size=50, generations=generations, mutation_rate=mutation_rate, crossover_rate=crossover_rate)
        plt.plot(fitness_history, label=f'crossover_rate={crossover_rate}')
    
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) tt nht')
    plt.title('nh hng ca t l lai ghp n tc  hi t')
    plt.legend()
    plt.show()

# Chy phn tch
analyze_parameters()
````

## Mc tiu

- So snh hiu qu ca phng php la chn mi vi phng php c (v d: Chn ngu nhin, Tournament Selection, v.v.).
- Phn tch nh hng ca phng php la chn n tc  hi t v cht lng nghim.

---

##  Cc phng php la chn ph bin

| Phng php                | M t                                                                                   | u im                              | Nhc im                          |
|----------------------------|-----------------------------------------------------------------------------------------|------------------------------------|-----------------------------------|
| Chn ngu nhin (Random)   | Chn c th ngu nhin trong qun th                                                   | n gin, nhanh                    | Khng u tin c th tt hn      |
| Roulette Wheel Selection    | Chn theo xc sut t l thun vi  thch nghi (fitness)                              | D dng, u tin c th tt        | D b nh hng bi c th qu mnh|
| Tournament Selection        | Chn ngu nhin mt s c th, chn c th tt nht trong nhm                          | Kim sot a dng tt              | Cn chn kch thc nhm ph hp  |

---

##   tng ci tin

1. Ci t thm phng php **Roulette Wheel Selection** trong thut ton di truyn.
2. Chy th vi cng b tham s trn cc phng php la chn khc nhau.
3. So snh kt qu v:
   - Tc  hi t (s th h cn  t nghim tt).
   - Gi tr nghim tt nht t c.
4. Phn tch u  nhc im tng phng php trong thc t.

---

##  Kt qu v phn tch

- Phng php **Roulette Wheel** thng gip tng tc  hi t nh u tin c th tt.
- Tuy nhin c th lm gim a dng gen nu c th u t chim u th qu ln, d ri vo cc tr cc b.
- Phng php **Tournament** kim sot tt s a dng nhng c th chm hn trong mt s trng hp.
- Phng php **Random** n gin nhng t hiu qu trong ti u nng cao.

---


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def target_function(x):
    return np.sin(x) + np.cos(x)

# Tnh fitness - m bo lun dng bng cch cng thm 2
def fitness_function(x):
    return target_function(x) + 2

# M ha: Chuyn i gi tr thc thnh chui nh phn
def encode(x, bits=16, min_val=0, max_val=2 * np.pi):
    normalized_value = (x - min_val) / (max_val - min_val)
    integer_value = int(normalized_value * (2**bits - 1))
    return format(integer_value, f'0{bits}b')

# Gii m: Chui nh phn -> gi tr thc
def decode(binary_string, bits=16, min_val=0, max_val=2 * np.pi):
    integer_value = int(binary_string, 2)
    normalized_value = integer_value / (2**bits - 1)
    return min_val + normalized_value * (max_val - min_val)

# Tnh ton fitness cho qun th
def calculate_fitness(population):
    return [fitness_function(decode(ind)) for ind in population]

# Chn lc ngu nhin
def random_selection(population):
    return random.choice(population), random.choice(population)

# Chn lc (Roulette Wheel Selection)
def roulette_wheel_selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness <= 0:
        return random_selection(population)
    
    selection_probs = [f / total_fitness for f in fitness_values]
    return np.random.choice(population, size=2, p=selection_probs)

# Chn lc (Tournament Selection)
def tournament_selection(population, fitness_values, tournament_size=3):
    selected_indices = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected_indices, key=lambda i: fitness_values[i])
    return population[best_idx], population[best_idx]

# Lai ghp 1 im vi xc sut crossover_rate
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        child = parent1[:point] + parent2[point:]
        return child
    return parent1

# t bin vi xc sut mutation_rate
def mutate(individual, mutation_rate=0.05):
    return ''.join(
        bit if random.random() > mutation_rate 
        else '1' if bit == '0' else '0' 
        for bit in individual
    )

# Thut ton di truyn chnh
def genetic_algorithm(selection_method, pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8, bits=16, min_val=0, max_val=2*np.pi):
    population = [encode(random.uniform(min_val, max_val), bits) for _ in range(pop_size)]
    best_fitness_history = []

    for gen in range(generations):
        fitness_values = calculate_fitness(population)
        best_target = max(fitness_values)
        best_fitness_history.append(best_target)

        new_population = []
        for _ in range(pop_size):
            if selection_method == 'random':
                parent1, parent2 = random_selection(population)
            elif selection_method == 'roulette':
                parent1, parent2 = roulette_wheel_selection(population, fitness_values)
            elif selection_method == 'tournament':
                parent1, parent2 = tournament_selection(population, fitness_values)
            else:
                raise ValueError("Unknown selection method")

            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate)
            new_population.append(child)

        population = new_population

        # Hin th kt qu tt nht mi th h
        best_idx = np.argmax(fitness_values)
        best_x = decode(population[best_idx])
        print(f"Gen {gen + 1}: x = {best_x:.4f}, h(x) = {best_target:.4f}")

    return best_fitness_history


# Phn tch hiu qu ca cc phng php la chn
def compare_selection_methods():
    methods = ['random', 'roulette', 'tournament']
    results = {}
    
    for method in methods:
        print(f"\nChy vi phng php la chn: {method}")
        fitness_history = genetic_algorithm(selection_method=method, pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8)
        results[method] = fitness_history
        plt.plot(fitness_history, label=f'Selection: {method}')
    
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) tt nht')
    plt.title('So snh cc phng php la chn trong thut ton di truyn')
    plt.legend()
    plt.show()

# Chy phn tch
compare_selection_methods()
                
````

- Dng biu  phn tn (scatter plot)  th hin v tr cc c th trn khng gian bin s qua tng th h.

##  tng thc hin

- Vi bi ton ti u hm $k(x, y, z) = x^2 + y^2 + z^2$, khng gian tm kim l 3 chiu.
- Mi c th l mt im trong khng gian $(x, y, z)$.
-  mi th h, ly v tr gii m ca ton b c th, v scatter plot.
- Nu khng th trc quan 3D trc tip, c th v scatter 2D tng cp bin: $(x, y)$, $(y, z)$, $(x, z)$ qua cc th h.

````python
#code vo y
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def target_function(variables):
    return sum(x**2 for x in variables)

# Tnh fitness - m bo lun dng
def fitness_function(variables):
    return -target_function(variables)  # Ti u ha hm k(x,y,z)

# M ha: Chuyn i gi tr thc thnh chui nh phn
def encode(variables, bits=16, min_val=-10, max_val=10):
    encoded = []
    for var in variables:
        normalized_value = (var - min_val) / (max_val - min_val)
        integer_value = int(normalized_value * (2**bits - 1))
        encoded.append(format(integer_value, f'0{bits}b'))
    return ''.join(encoded)  # Tr v chui nh phn duy nht

# Gii m: Chui nh phn -> gi tr thc
def decode(encoded_str, bits=16, min_val=-10, max_val=10):
    decoded = []
    for i in range(0, len(encoded_str), bits):
        binary_string = encoded_str[i:i + bits]
        integer_value = int(binary_string, 2)
        normalized_value = integer_value / (2**bits - 1)
        decoded.append(min_val + normalized_value * (max_val - min_val))
    return decoded

# Tnh ton fitness cho qun th
def calculate_fitness(population):
    return [fitness_function(decode(ind)) for ind in population]

# Chn lc (Roulette Wheel Selection)
def roulette_wheel_selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness <= 0:
        return random.choice(population), random.choice(population)
    
    selection_probs = [f / total_fitness for f in fitness_values]
    return np.random.choice(population, size=2, p=selection_probs)

# Lai ghp 1 im vi xc sut crossover_rate
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        child = parent1[:point] + parent2[point:]
        return child
    return parent1

# t bin vi xc sut mutation_rate
def mutate(individual, mutation_rate=0.05):
    return ''.join(
        bit if random.random() > mutation_rate 
        else '1' if bit == '0' else '0' 
        for bit in individual
    )

# Thut ton di truyn chnh
def genetic_algorithm(pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8, bits=16, min_val=-10, max_val=10):
    population = [encode([random.uniform(min_val, max_val) for _ in range(3)], bits, min_val, max_val) for _ in range(pop_size)]
    best_fitness_history = []
    all_population_history = []  # Lu tr qun th qua cc th h
    
    for gen in range(generations):
        fitness_values = calculate_fitness(population)
        best_target = max(fitness_values)
        best_fitness_history.append(best_target)
        
        # Lu tr qun th hin ti
        all_population_history.append([decode(ind) for ind in population])
        
        new_population = []
        for _ in range(pop_size):
            parent1, parent2 = roulette_wheel_selection(population, fitness_values)
            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate)
            new_population.append(child)
        
        population = new_population
        
        # Hin th kt qu tt nht mi th h
        best_idx = np.argmax(fitness_values)
        best_vars = decode(population[best_idx])
        print(f"Gen {gen + 1}: x = {best_vars[0]:.4f}, y = {best_vars[1]:.4f}, z = {best_vars[2]:.4f}, fitness = {best_target:.4f}")
    
    return best_fitness_history, all_population_history

# Trc quan ha qun th
def visualize_population(all_population_history, step=10):
    for gen in range(0, len(all_population_history), step):
        population = np.array(all_population_history[gen])
        
        # V scatter plot cho tng cp bin
        plt.figure(figsize=(12, 8))
        plt.scatter(population[:, 0], population[:, 1], label='(x, y)', alpha=0.5)
        plt.scatter(population[:, 1], population[:, 2], label='(y, z)', alpha=0.5)
        plt.scatter(population[:, 0], population[:, 2], label='(x, z)', alpha=0.5)
        plt.title(f'Population Distribution at Generation {gen + 1}')
        plt.xlabel('Variable 1')
        plt.ylabel('Variable 2')
        plt.legend()
        plt.grid()
        plt.show()

# Chy thut ton v trc quan ha
best_fitness_history, all_population_history = genetic_algorithm()
visualize_population(all_population_history)
````
